# Basic Math for ML(1)
## 1.함수
- 두 집합 사이의 관계 혹은 규칙
- y = f(x) 형태로 표현, x는 입력 값, y는 출력 값
- 데이터의 함수에서 x는 feature, y는 label로 둘 사이의 관계를 표현하는 것
- 햇(^)을 씌울 경우 예측 값

### 1-1. 일차 함수
- y가 x에 대한 일차식으로 표현된 경우
- y = ax + b (a != 0)
- a를 기울기, b를 절편이라고 표현

### 1-2. 이차 함수
- y과 x에 대한 이차식으로 표현된 경우
- y = a(x-p)2+q (a != 0)

## 2. 순간 변화율(접선의 기울기)
- x의 값이 미세하게 변화했을 때 y의 변화율

## 3. 미분
- 함수 f(x)를 미분한다는 것은 함수의 순간 변화율을 구한다는 것

## 4. 함수의 최솟값
- 함수의 최솟값에서의 미분값은 항상 0
- 이를 바탕으로 파라미터의 최적값을 구할 수 있음
